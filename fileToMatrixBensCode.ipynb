```python
import pandas as pd
import itertools
import numpy as np
import matplotlib.pyplot as plt
    
def build_matrix(entireDataframe):
    
    correction = 1000000
    chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX"]
    chromosomeLength = [(chrom, 0) for chrom in chromosomes]
    blockMatrix = [[0 for _ in range(23)] for _ in range(23)]



    for _, row in entireDataframe.iterrows():
        m = chromosomes.index(row[2])
        if row[3] > chromosomeLength[m][1]:
            chromosomeLength[m] = (chromosomeLength[m][0], row[3])

    
    chromosome_lengths = [
    ["chr1", 249],
    ["chr2", 243],
    ["chr3", 197],
    ["chr4", 191],
    ["chr5", 180],
    ["chr6", 171],
    ["chr7", 159],
    ["chr8", 146],
    ["chr9", 141],
    ["chr10", 135],
    ["chr11", 134],
    ["chr12", 133],
    ["chr13", 115],
    ["chr14", 107],
    ["chr15", 102],
    ["chr16", 90],
    ["chr17", 81],
    ["chr18", 78],
    ["chr19", 59],
    ["chr20", 62],
    ["chr21", 48],
    ["chr22", 51],
    ["chrX", 154]
]


    for chrom, length in chromosome_lengths:
        
        size = length
        k = chromosomes.index(chrom)
        subMatrix = [[0 for _ in range(size+1)] for _ in range(size+1)]
        blockMatrix[k][k] = subMatrix

    sum = 0

    for _, row in entireDataframe.iterrows():

        k = chromosomes.index(row[0])
        #l = chromosomes.index(row[2])
        i = int(row[1] // correction)
        j = int(row[3] // correction)

        size = chromosomeLength[k][1]


        m = chromosomes.index(row[2])
        chromosomeLength[m][1]
        
        submatrix = blockMatrix[k][k]
       
        if 0 <= i < size and 0 <= j < size:
            submatrix[i][j] = row[4]
            submatrix[j][i] = row[4]
            sum = sum + 1
   
    return blockMatrix, sum

```

```python
## Mean, variance

summary_statistics = []
chromosomes = [f"chr{i}" for i in range(1, 23)] + ["chrX"]


for i in range(4):
    for cell_id, matrix in data_dict[directory[i]].items():
        for j in range(23):
            chromosome_matrix = matrix[j][j]
            diagonal = np.diag(chromosome_matrix)
            to_process = np.array(chromosome_matrix)

            


            N = to_process.shape[0]
            # Count nonzero entries in the upper triangle, excluding the diagonal
            E = np.count_nonzero(np.triu(to_process, k=1))

            # Compute density
            density = (2 * E) / (N * (N - 1))
            
            sum_of_contacts = np.sum(to_process) / N
            variance_of_contacts = np.var(to_process)
            total_zero_count = np.count_nonzero(to_process == 0)
            nonzero_element_count =int((len(chromosome_matrix))**2 - total_zero_count) / N
            total_nonzero_ratio = 1 - ( total_zero_count / len(chromosome_matrix)**2 ) 
            zero_along_diagonal_count = np.count_nonzero(diagonal == 0) / N
            zero_ratio_along_diagonal = zero_along_diagonal_count/len(diagonal)
            variance_along_diagonal  = np.var(diagonal)
            variance_off_diagonal = np.var(to_process[~np.eye(to_process.shape[0], dtype=bool)])
            entropy_val = shannon_entropy(chromosome_matrix) / N
            trace = np.sum(diagonal) / N


            summary_statistics.append([
                directory[i], 
                chromosomes[j], 
                sum_of_contacts, 
                trace, 
                variance_of_contacts,  
                variance_along_diagonal,
                variance_off_diagonal,
                density, 
                total_nonzero_ratio, 
                zero_ratio_along_diagonal, 
                entropy_val])

summary_stats_dataframe = pd.DataFrame(summary_statistics, columns=[
    "cell_type",
    "chromosome",
    "sum",
    "trace",
    "variance",
    "variance_along_diagonal",
    "variance_off_diagonal",
    "density",
    "zero_ratio",
    "diagonal_zero_ratio",
    "entropy"
])

# List of metrics to plot
metrics = [
    "sum",
    "trace",
    "variance",
    "variance_along_diagonal",
    "variance_off_diagonal",
    "density",
    "zero_ratio",
    "diagonal_zero_ratio",
    "entropy"
]

# Enforce chromosome order
chrom_order = [f"chr{i}" for i in range(1, 23)] + ["chrX"]

# For each metric, compute mean per chromosome and cell type
for metric in metrics:
    grouped = (
        summary_stats_dataframe
        .groupby(["chromosome", "cell_type"])[metric]
        .mean()
        .unstack()  # rows: chromosomes, columns: cell types
        .reindex(chrom_order)  # ensure chromosome order
    )

    # Plot
    grouped.plot(kind="bar", figsize=(14, 6))
    plt.title(f"Mean {metric.replace('_', ' ').title()} per Chromosome (All Cell Types)")
    plt.xlabel("Chromosome")
    plt.ylabel(f"Mean {metric.replace('_', ' ').title()}")
    plt.xticks(rotation=45)
    plt.legend(title="Cell Type")
    plt.grid(axis="y")
    plt.tight_layout()
    plt.show()

"""

summary_stats_dataframe = pd.DataFrame(summary_statistics, columns=[
    "cell_type",
    "chromosome",
    "sum",
    "variance",
    "variance_along_diagonal",
    "zero_ratio",
    "diagonal_zero_ratio",
    "entropy"
])

# Set chromosome as index
df_by_chr = summary_stats_dataframe.set_index("chromosome")

# List of metrics to plot
metrics = [
    "sum",
    "variance",
    "variance_along_diagonal",
    "zero_ratio",
    "diagonal_zero_ratio",
    "entropy"
]

mean_per_chr_stats = {
    "sum": summary_stats_dataframe.groupby("chromosome")["sum"].mean(),
    "variance": summary_stats_dataframe.groupby("chromosome")["variance"].mean(),
    "variance_along_diagonal": summary_stats_dataframe.groupby("chromosome")["variance_along_diagonal"].mean(),
    "zero_ratio": summary_stats_dataframe.groupby("chromosome")["zero_ratio"].mean(),
    "diagonal_zero_ratio": summary_stats_dataframe.groupby("chromosome")["diagonal_zero_ratio"].mean(),
    "entropy": summary_stats_dataframe.groupby("chromosome")["entropy"].mean()
}


# Optional: consistent chromosome order
chrom_order = [f"chr{i}" for i in range(1, 23)] + ["chrX"]
for key in mean_per_chr_stats:
    mean_per_chr_stats[key] = mean_per_chr_stats[key].reindex(chrom_order)

for metric, series in mean_per_chr_stats.items():
    plt.figure(figsize=(12, 6))
    series.plot(kind="bar")
    plt.title(f"Mean {metric.replace('_', ' ').title()} per Chromosome (GM12878)")
    plt.xlabel("Chromosome")
    plt.ylabel(f"Mean {metric.replace('_', ' ').title()}")
    plt.xticks(rotation=45)
    plt.grid(axis="y")
    plt.tight_layout()
    plt.show()




# Loop through each metric and plot separately
for metric in metrics:
    plt.figure(figsize=(10, 4))
    df_by_chr[metric].plot(kind="bar")
    plt.title(f"{metric.replace('_', ' ').title()} per Chromosome")
    plt.xlabel("Chromosome")
    plt.ylabel(metric.replace('_', ' ').title())
    plt.xticks(rotation=90)
    plt.grid(True, axis='y')
    plt.tight_layout()
    plt.show()



"""


```

```python
contact_distance_data = []

for i in range(4):
    cell_type = directory[i]
    for cell_id, matrix in data_dict[cell_type].items():
        for j in range(23):
            chromosome_matrix = np.array(matrix[j][j])  # Ensure it's a NumPy array
            chromosome = chromosomes[j]

            n = chromosome_matrix.shape[0]
            for row in range(n):
                for col in range(n):
                    contact = chromosome_matrix[row, col]
                    if contact != 0:
                        distance = abs(row - col)
                        contact_distance_data.append([
                            distance, contact, cell_type, chromosome
                        ])

# Convert to DataFrame
df_contacts = pd.DataFrame(contact_distance_data, columns=[
    "distance", "contact", "cell_type", "chromosome"
])

# ðŸŽ» Violin plots: 1 per cell type, with 23 violins (1 per chromosome)
for cell_type in directory:
    plt.figure(figsize=(16, 6))
    subset = df_contacts[df_contacts["cell_type"] == cell_type]
    sns.violinplot(
        data=subset,
        x="chromosome",
        y="contact",
        palette="Set2",
        inner="box",
        cut=0
    )
    plt.title(f"Contact Distribution vs. Distance by Chromosome ({cell_type})")
    plt.xlabel("Chromosome")
    plt.ylabel("Contact Frequency")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
