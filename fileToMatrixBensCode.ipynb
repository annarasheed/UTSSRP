```python

import os
import random
import glob


GM12878sample = []


# Set the path to the folder containing the text files
folder_name = "GM12878"  # change this to your folder name
folder_path = os.path.join(os.getcwd(), folder_name)

# Get a list of all .txt files in that folder
txt_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Select 15 random files, or fewer if there are less than 15
selected_files = random.sample(txt_files, min(20, len(txt_files)))

# Print or process the selected files
for filename in selected_files:

    if filename.endswith(".txt"):  # Adjust for your actual file type
        filepath = filename
        print(filename)
        cell_type = "GM12878"
        cell_id = filename  
        dataframe = pd.read_csv(filepath, sep='\t', header=None)
        matrix = build_matrix(dataframe)
        GM12878sample.append([matrix, cell_type, filename, 0])





HAP1sample = []

# Set the path to the folder containing the text files
folder_name = "HAP1"  # change this to your folder name
folder_path = os.path.join(os.getcwd(), folder_name)

# Get a list of all .txt files in that folder
txt_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Select 15 random files, or fewer if there are less than 15
selected_files = random.sample(txt_files, min(20, len(txt_files)))

# Print or process the selected files
for filename in selected_files:

    if filename.endswith(".txt"):  # Adjust for your actual file type
        filepath = filename
        cell_type = "HAP1"
        cell_id = filename  
        dataframe = pd.read_csv(filepath, sep='\t', header=None)
        matrix = build_matrix(dataframe)
        HAP1sample.append([matrix, cell_type, filename, 0])




Helasample = []

# Set the path to the folder containing the text files
folder_name = "Hela"  # change this to your folder name
folder_path = os.path.join(os.getcwd(), folder_name)

# Get a list of all .txt files in that folder
txt_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Select 15 random files, or fewer if there are less than 15
selected_files = random.sample(txt_files, min(20, len(txt_files)))

# Print or process the selected files
for filename in selected_files:

    if filename.endswith(".txt"):  # Adjust for your actual file type
        filepath = filename
        cell_type = "Hela"
        cell_id = filename  
        dataframe = pd.read_csv(filepath, sep='\t', header=None)
        matrix = build_matrix(dataframe)
        Helasample.append([matrix, cell_type, filename, 0])




K562sample = []

# Set the path to the folder containing the text files
folder_name = "K562"  # change this to your folder name
folder_path = os.path.join(os.getcwd(), folder_name)

# Get a list of all .txt files in that folder
txt_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Select 15 random files, or fewer if there are less than 15
selected_files = random.sample(txt_files, min(20, len(txt_files)))

# Print or process the selected files
for filename in selected_files:

    if filename.endswith(".txt"):  # Adjust for your actual file type
        filepath = filename
        cell_type = "K562"
        cell_id = filename  
        dataframe = pd.read_csv(filepath, sep='\t', header=None)
        matrix = build_matrix(dataframe)
        K562sample.append([matrix, cell_type, filename, 0])






from scipy.stats import entropy

def shannon_entropy(matrix, base=2):
    """
    Compute Shannon entropy of a contact matrix.
    The matrix should be non-negative and not all zeros.
    """
    matrix = np.array(matrix, dtype=np.float64)
    
    # Flatten the matrix and remove zero entries (optional)
    flat = matrix.flatten()
    flat = flat[flat > 0]

    # Normalize to create a probability distribution
    probs = flat / flat.sum()

    # Calculate entropy
    return entropy(probs, base=base)



allSamples = [K562sample, Helasample, GM12878sample,HAP1sample]

for sampleGroup in allSamples:
    for i in range(len(sampleGroup)):
        matrix, cell_type, filename, _ = sampleGroup[i]
        chromosomeEntropy = []
        
        for j in range(len(matrix)):
            chrom_contact_matrix = matrix[j][j]  # This should be a 2D array
            entropy_val = shannon_entropy(chrom_contact_matrix)
            statEntry = [cell_type, filename, j, entropy_val]
            chromosomeEntropy.append(statEntry)

        # Save the list of per-chromosome entropy results
        sampleGroup[i][3] = chromosomeEntropy


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Flatten all entropy results into a single list
entropy_records = []
for sampleGroup in allSamples:
    for _, cell_type, filename, chrom_stats in sampleGroup:
        for stat in chrom_stats:
            stat_cell_type, stat_filename, chrom_index, entropy_val = stat
            entropy_records.append({
                'CellType': stat_cell_type,
                'Filename': stat_filename,
                'Chromosome': f"Chr{chrom_index+1}",
                'Entropy': entropy_val
            })

            
# Convert to DataFrame
df_entropy = pd.DataFrame(entropy_records)
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

# Dot plot: Entropy vs. Chromosome, color-coded by CellType
sns.stripplot(data=df_entropy, 
              x='Chromosome', 
              y='Entropy', 
              hue='CellType', 
              dodge=True, 
              jitter=False, 
              palette='Set2')






plt.title("Per-Chromosome Contact Entropy Across Cell Types")
plt.xticks(rotation=45)
plt.legend(title="Cell Type")
plt.tight_layout()
plt.show()
```
